**状態方程式**
$$
\begin{align}
x_t &= β_t * x_{t-1} + ν_t \\
y_t &= e ^ {(α_t + x_t)/2} * w_t \\
\end{align}
$$

$$
w_t \sim \rm{Normal(0,1)}, \nu_t \sim Normal(0, \tau^2)
$$

```julia
# 粒子フィルタ
function particle_filter(
    y;
    targets=[:y],
    states=[:x], 
    params=[:α, :β, :τ],
    n_particles=10000, 
    init=(x=Normal(0,1), α=Normal(10,1), β=Uniform(0,1), τ=Normal(-1,1)),
    f = (x,β,d) -> β * x + rand(d),
    h = (x, α) -> exp((x + α)/2),
    loglikelihood = (v,d) -> loglikelihood(d, v)
)
    labels = [targets..., states..., params...]
    @assert length(labels) == length(unique(labels))
    n_obs_dim = size(y,2)
    n_particle_dim = length(states) + length(params) # 粒子数
    particles = zeros(n_particles, n_particle_dim)   # 粒子配列
    predicted = zeros(n_particles, n_obs_dim)   # 予測粒子
    weights = zeros(n_particles) # 粒子重み
    loglik = zeros(size(y,1))

    # 各時刻の粒子の統計量を保存する変数
    stats = OrderedDict{Symbol, DataFrame}(l => DataFrame() for l in labels)

    # 粒子初期化
    for (i, symbol) in enumerate(states)
        d = getproperty(init, symbol)
        particles[:,i] = rand(d, n_particles)
    end
    for (i, symbol) in enumerate(params)
        d = getproperty(init, symbol)
        particles[:,i+length(states)] = rand(d, n_particles)
    end
    
    for i in axes(y,1)
        # 状態遷移
        for j in axes(particles, 2)
            j > 2 && break
            for k in axes(particles, 1)
                τ = sqrt(exp(particles[k,4])) + 1e-9
                x = particles[k,1]
                β = particles[k,3]
                particles[k,1] = f(x, β, Normal(0, τ))
            end
        end
        
        # 粒子重み計算
        for k in eachindex(weights)
            x = particles[k,1]
            α = particles[k,2]
            predicted[k] = σ = h(x, α)
            weights[k] = loglikelihood(y[i], Normal(0, σ))
        end
        loglik[i] = mean(weights)
        weights .= exp.(weights .- maximum(weights))
        weights ./= sum(weights)

        # 統計量保存
        add_stats!(stats, predicted, :y, 1)
        for (index, symbol) in enumerate([:x, :α, :β, :τ])
            add_stats!(stats, particles, symbol, index)
        end
        
        # リサンプリング
        new_indices = rand(Categorical(weights), n_particles)
        particles = particles[new_indices,:]
    end
    stats, loglik
end

# 各粒子の統計量を算出する
function add_stats!(stats, particles, label, index)
    push!(stats[label], (
        median=median(particles[:,index]),
        q25=quantile(particles[:,index], 0.25),
        q75=quantile(particles[:,index], 0.75),
    ))
end

# 価格差をモデリングする
# TODO: 定常状態になるまでwamup期間を考慮する
df = dropmissing(df)
ts = df[:, :timestamp]
y = df[:, :AdjClose]
y_diff = diff(y)
results, loglik = particle_filter(y_diff)
@show mean(loglik)

n = length(keys(results))
p = plot(layout=(n,1), size=(900,600))
skip = 20
for (i, key) in enumerate(keys(results))
    !hasproperty(results[key], :median) && continue
    plot!(p[i], results[key].median[skip:end], label=string(key),
        fillrange = (results[key].q25[skip:end], results[key].q75[skip:end]),
        fillalpha = 0.3
    )
end
display(p)

# ribbon = (results[:α].q75 .- results[:α].q25)
# plot(results[:α].mean, ribbon=ribbon) |> display

# ribbon = (results[:β].q75 .- results[:β].q25)
# plot(results[:β].mean, ribbon=ribbon) |> display

# ribbon = (results[:τ].q75 .- results[:τ].q25)
# plot(results[:τ].mean, ribbon=ribbon)

# vo = exp.((results[:x].mean .+ results[:α].mean) ./ 2)
# plot(vo)
```
