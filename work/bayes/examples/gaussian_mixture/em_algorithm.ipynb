# 混合ガウスモデルのEMアルゴリズム実装

```julia
include("./models.jl")

model = example_model()
dataset = make_dataset(model, 100)
```

```julia
function scatter_plot!(xs, ys)
    for (i,d) in enumerate(model.components)
        idxs = findall(ys .== i)
        scatter!(xs[1,idxs], xs[2,idxs], color=i, colormap=:tab10, colorrange=(1,10))
    end
end

# # 関数値を取得

f = Figure()
Axis(f[1, 1])
contour_plot!(model)
scatter_plot!(dataset...)
# データプロット

f
```

```julia
include("em_algorithm.jl")
xs, ys = dataset
n_components = 3

result = fit(EMAlgorithm(), xs=xs, n_components=n_components, max_iter=50)
```

```julia
# 結果表示
f = Figure()
Axis(f[1,1])
contour_plot!(result)
scatter_plot!(dataset...)
f
```
